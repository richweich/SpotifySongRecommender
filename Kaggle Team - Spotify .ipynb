# Model that takes in Songs already in Dataset

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(recommender_col2[columns_to_scale], recommender_col2[columns_to_scale])

def recommend_similar_songs(song_index, top_n=10):
    # Get similarity scores for the input song
    song_similarities = similarity_matrix[song_index]
    # Get indices of top N similar songs
    top_indices = song_similarities.argsort()[-top_n-1:-1][::-1]  # Exclude the input song itself
    return top_indices

def print_recommended_songs(song_name, top_n=10):
    # Find the index of the input song
    song_indices = df2.index[df2['track_name'] == song_name].tolist()
    if not song_indices:
        print("Song '{}' not found in the dataset.".format(song_name))
        return []
    song_index = song_indices[0]
    # Get indices of top N similar songs
    similar_song_indices = recommend_similar_songs(song_index, top_n)
    # Filter out invalid indices and limit to top_n
    valid_indices = [i for i in similar_song_indices if i in df2.index][:top_n]
    # Get track names and artist names of top N similar songs
    similar_songs_info = [(df2.loc[i, 'track_name'], df2.loc[i, 'artist(s)_name']) for i in valid_indices]
    return similar_songs_info

while True:
    # User input
    track_name = input("Enter the name of the track (type 'exit' to quit): ")
    
    # Check if the user wants to exit
    if track_name.lower() == 'exit':
        print("Exited")
        break
    
    # Search for the index of the track name
    matching_indices = df[df["track_name"] == track_name].index
    if len(matching_indices) == 0:
        print("Song not found. Try again.")
    else:
        track_index = matching_indices[0]
        recommendations = print_recommended_songs(track_name)
        if recommendations:
            print("Recommended songs for '{}':".format(track_name))
            for track_name, artist_name in recommendations:
                print("Song: {}\nArtist: {}\n".format(track_name, artist_name))
        else:
            print("No similar songs found.")
```
